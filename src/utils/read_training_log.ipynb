```python
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Open and read the JSON file
with open('../../data/models/batch_128_15625_500000/trainer_state.json', 'r') as file:
    data = json.load(file)['log_history']
```

```python
train_loss_y = [data[val]['loss'] for val in range(len(data)) if 'loss' in data[val]]
train_loss_x = [data[val]['step'] for val in range(len(data)) if 'loss' in data[val]]
eval_loss_y = [data[val]['eval_loss'] for val in range(len(data)) if 'eval_loss' in data[val]]
eval_loss_x = [data[val]['step'] for val in range(len(data)) if 'eval_loss' in data[val]]
```

```python
plt.plot(train_loss_x, train_loss_y, linewidth=2)
plt.plot(eval_loss_x, eval_loss_y, linewidth=2, marker='o')
plt.xlabel("Step (batch size = 128)")
plt.ylabel("Loss")
plt.legend(["Training loss", "Eval loss"])
```
